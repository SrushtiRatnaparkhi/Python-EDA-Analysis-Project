## Exploratory Data **Analysis**

Gender

```python
#plotting a bar chart for Gender and it's count
sns.set(rc={'figure.figsize':(10,5)})
palette = sns.color_palette("husl", len(df['Gender'].unique()))
gender = sns.countplot(x='Gender', data=df,  hue='Gender',palette=palette)
for bars in gender.containers:
    gender.bar_label(bars)
```

```python

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
ax = sns.barplot(x='Gender', y='Amount',data=sales_gen, hue='Gender', palette=palette, dodge=False)
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

### **AGE**

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
ax = sns.barplot(x='Age Group', y='Amount', data=sales_age, hue='Age Group', palette='viridis', dodge=False)
ax.set_xlabel('Age Group')
```

From above graphs we can see that most of the buyers are of age group between 26-35 yrs *female*

## **State**

```python

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})

# Define the color palette
palette = sns.color_palette("husl", len(sales_state))

# Create a bar plot with hue set to 'State' and legend set to False
ax = sns.barplot(data=sales_state, x='State', y='Orders', hue='State', palette=palette, dodge=False)

```

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
palette = sns.color_palette("husl", len(sales_state))
ax = sns.barplot(data=sales_state, x='State', y='Amount', hue='State', palette=palette, dodge=False)



```

From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively

# **Marital Status**

```python
palette = sns.color_palette("husl", len(df['Marital_Status'].unique()))
sns.set(rc={'figure.figsize':(7,5)})

ax = sns.countplot(data=df, x='Marital_Status', hue='Marital_Status', palette=palette, dodge=False)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```



From above graphs we can see that most of the buyers are married (women) and they have high purchasing power



Occupation

```python

sns.set(rc={'figure.figsize':(25,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
counts = df['Occupation'].value_counts()
palette = sns.color_palette("husl", len(df['Occupation'].unique()))
ax = sns.countplot(data=df, x='Occupation', order=counts.index, hue='Occupation', palette=palette)
```

```python
sns.set(rc={'figure.figsize':(25,5)})
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

ax = sns.barplot(data=sales_state, x='Occupation', y='Amount', hue='Occupation', palette=palette, dodge=False)

```

From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

Product Category



```python
sns.set(rc={'figure.figsize':(25,5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
counts = df['Product_Category'].value_counts().sort_values(ascending=False)
sns.set(rc={'figure.figsize': (40, 5)})
palette = sns.color_palette("husl", len(counts))
ax = sns.countplot(data=df, x='Product_Category', order=counts.index, hue='Product_Category', palette=palette)
ax.set_xlabel('Product Category')
ax.set_ylabel('Count')
```

From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
palette = sns.color_palette("husl", len(sales_state))
ax = sns.barplot(data=sales_state, x='Product_ID', y='Orders', palette=palette, hue='Product_ID')

```

```python
# top 10 most sold products (same thing as above)

fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
